```python
%%vspipe --y4m | ffmpeg -r 24000/1001*2 -y -i "video/test.mp4" -i - -map 0:a -map 1:v -vcodec hevc_nvenc -c:a copy -qp 16 -preset 18 -profile:v 1 out_2x.mkv
from vsdrba_distilled import drba_distilled
import vapoursynth as vs
from vapoursynth import core
clip = core.bs.VideoSource(source="video/test.mp4")
clip = core.resize.Bicubic(clip=clip, matrix_in_s="709", format=vs.RGBH,width=1920, height=1080)
ret = drba_distilled(clip, trt=True, factor_num=2, factor_den=1, scale=1.0, model="v2_lite", auto_download=True)
clip = core.resize.Bicubic(clip=ret, matrix_s="709", format=vs.YUV420P16)
clip.set_output()
```
